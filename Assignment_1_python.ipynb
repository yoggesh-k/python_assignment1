```python
# Ques3: Compare and contrast mutable and immutable objects in Python with examples
 
 
 ''' Ans:  Python, objects can be categorized as either mutable or immutable:
 
 1. Mutable: Mutable objects allow changes after creation.That means Mutable objects can be changed after they 
     are created.This means we can alter their state or contents without creating a new object.
 Common examples of mutable objects in Python include lists, dictionaries, sets, and most user-defined objects.
     
     Characteristics of Mutable Objects:
     
      1. Contents can change : The elements or attributes of the object can be modified.
      2. Same memory location : When modified, the object retains its memory address.

    
      2. Immutable: Immutable objects cannot be changed after they are created.That means Immutable objects 
      cannot be changed after they are created. Any attempt to modify the object results in the creation of a
      new object. Common examples of immutable objects in Python include integers, floats, strings, tuples.
    
    Characteristics of Immutable Objects:
    
     1. Contents cannot change: Once created, the object's state or content cannot be altered.
     2. New memory location: Modifying an immutable object results in the creation of a new object at different 
        memory address.

     Comparison: (Mutable and Immutable Objects)
     
   1. Similarities:
       Both types of objects are used to store data.
       
       Both can be assigned to variables and passed to functions.

   2. Differences:
   
       Modification:
        Mutable objects can be modified in place without changing their memory address.
        
        Immutable objects cannot be modified in place; any change results in a new object.

   3.Performance:
       Mutable objects can be more efficient for operations that involve many changes, as they do not require
       creating new objects.
       
       Immutable objects can be more efficient in some contexts due to their hashability, making them suitable
       for use as dictionary keys or set elements.

    4.Use Cases:
        Mutable objects are often used when a data structure needs to be frequently updated, such as lists for
        dynamic collections of items.
        
        Immutable objects are often used for fixed data structures where safety from unintended changes is 
        important, such as tuples for fixed-size collections. '''
    
```

```python
#Example of Mutable objects:

#1.List:
my_list = [1, 2, 3]
print(my_list)  # Output: [1, 2, 3]

my_list.append(5)
print(my_list)  # Output: [1, 2, 3, 5]


#2. Dictionary:
my_dict = {'a': 1, 'b': 2}
print(my_dict)  # Output: {'a': 1, 'b': 2}

my_dict['c'] = 4
print(my_dict)  # Output: {'a': 1, 'b': 2, 'c': 4}
   

#3. Set
my_set = {1, 2, 3}
print(my_set)  # Output: {1, 2, 3}

my_set.add(10)
print(my_set)  # Output: {1, 2, 3, 10}
   
    
    
#Example of immutable objects:

#1. Integer:
x = 2
print(x)  # Output: 2

x = x + 1
print(x)  # Output: 3
   # Note: x now references a new integer object with value 3.

#2. String:
my_string = "hello"
print(my_string)  # Output: hello

new_string = my_string + " world"
print(new_string)  # Output: hello world
print(my_string)   # Output: hello
   # Note: new_string is a new string object, my_string remains unchanged.


#3. Tuple:
my_tuple = (1, 2, 3)
print(my_tuple)  # Output: (1, 2, 3)

new_tuple = my_tuple + (4,)
print(new_tuple)  # Output: (1, 2, 3, 4)
print(my_tuple)   # Output: (1, 2, 3)
   # Note: new_tuple is a new tuple object, my_tuple remains unchanged.
```

```python
#Ques4: Discuss the different types of operators in Python and provide examples of how they are used


''' Ans: In Python programming, Operators in general are used to perform operations on values and variables.
   These are standard symbols used for the purpose of logical and arithmetic operations.They can be categorized
   into several types based on their functionality.'''

     '''Here are the main types of operators in Python, along with examples of how they are used:                             
                                     (Given below in code) '''
```

```python
# 1. Arithmetic Operators

 #Arithmetic operators are used to perform mathematical operations.

# Addition (`+`)
x = 5
y = 3
result = x + y  # result is 8

# Subtraction (`-`)
  
result = x - y  # result is 2

# Multiplication (`*`) 
 

result = x * y  # result is 15

# Division (`/`) 
  
result = x / y  # result is 1.666..

# Floor Division (`//`)
 
result = x // y  # result is 1

# Modulus (`%`)
  
result = x % y  # result is 2


# Exponentiation (`**`)
  
result = x ** y  # result is 125


## 2. Comparison Operators

 #Comparison operators are used to compare two values.

#Equal (`==`)
 
result = (x == y)  # result is False


# Not Equal (`!=`)
  
result = (x != y)  # result is True
  

# Greater Than (`>`)
 
result = (x > y)  # result is True
  

# Less Than (`<`)
  
result = (x < y)  # result is False
  

# Greater Than or Equal To (`>=`)
  
result = (x >= y)  # result is True


# Less Than or Equal To (`<=`)
  
result = (x <= y)  # result is False
  

### 3. Logical Operators

# Logical operators are used to combine conditional statements.

# Logical AND (`and`)
  
result = (x > 2 and y < 5)  # result is True


# Logical OR (`or`)
 
result = (x > 2 or y > 5)  # result is True
  

# Logical NOT (`not`)
  
result = not(x > 2)  # result is False
  

### 4. Bitwise Operators

# Bitwise operators are used to perform bit-level operations on binary numbers.

# Bitwise AND (`&`)
  
result = x & y  # result is 1 (0101 & 0011 = 0001)
  

# Bitwise OR (`|`)
  
result = x | y  # result is 7 (0101 | 0011 = 0111)
  

# Bitwise XOR (`^`)
  
result = x ^ y  # result is 6 (0101 ^ 0011 = 0110)
  

# Bitwise NOT (`~`)
 
result = ~x  # result is -6 (~0101 = 1010)
  

# Bitwise Left Shift (`<<`)
  
result = x << 1  # result is 10 (0101 << 1 = 1010)
  

# Bitwise Right Shift (`>>`)
  
result = x >> 1  # result is 2 (0101 >> 1 = 0010)
  

### 5. Assignment Operators

# Assignment operators are used to assign values to variables.

# Assignment (`=`)
 
x = 5
  

# Add and Assign (`+=`)
  
x += 3  # equivalent to x = x + 3, x is now 8
  

# Subtract and Assign (`-=`)
  
x -= 2  # equivalent to x = x - 2, x is now 6


# Multiply and Assign (`*=`)
  
x *= 2  # equivalent to x = x * 2, x is now 12
  

# Divide and Assign (`/=`)
  
x /= 3  # equivalent to x = x / 3, x is now 4.0
  

# Floor Divide and Assign (`//=`)
  
x //= 2  # equivalent to x = x // 2, x is now 2.0
  

# Modulus and Assign (`%=`)
 

x %= 2  # equivalent to x = x % 2, x is now 0.0


# Exponentiate and Assign (`**=`)
  
x **= 3  # equivalent to x = x ** 3, x is now 0.0
  

# Bitwise AND and Assign (`&=`)
  
x = 5 

 
x &= 3  # equivalent to x = x & 3, x is now 1
  

# Bitwise OR and Assign (`|=`)
 
x |= 2  # equivalent to x = x | 2, x is now 3
  

# Bitwise XOR and Assign (`^=`)
  
x ^= 2  # equivalent to x = x ^ 2, x is now 1
  

# Bitwise Left Shift and Assign (`<<=`)
 
x <<= 1  # equivalent to x = x << 1, x is now 2
  

# Bitwise Right Shift and Assign (`>>=`)
  
x >>= 1  # equivalent to x = x >> 1, x is now 1
  

### 6. Identity Operators

 #Identity operators are used to compare the memory locations of two objects.

 # is
  
a = [1, 2, 3]
b = a
result = (a is b)  # result is True
  

# is not
  
c = a.copy()
result = (a is not c)  # result is True
  

# 7. Membership Operators

#Membership operators are used to test if a sequence contains an element.

# in
 
my_list = [1, 2, 3]
result = (2 in my_list)  # result is True
  

# not in
  
result = (4 not in my_list)  # result is True
  

### 8. Special Operators

"""Python has some special operators like the **walrus operator** (`:=`) introduced in Python 3.8, 
    which assigns values to variables as part of an expression."""

# Walrus Operator (`:=`)
  
if (n := len(my_list)) > 2:
    print(f"The list is too long ({n} elements)")
  
```

```python
# Ques6: How do conditional statements work in Python? Illustrate with examples

"""Ans:  Conditional statements in Python allow us to execute different blocks of code based on certain
         conditions. These statements use logical expressions to determine which code block should run.
         The primary conditional statements in Python are `if`, `elif`, and `else`. """
```

```python
#Examples 

#if Statement: The if statement executes a block of code if a condition is true.
if 10 > 5:
    print("10 is greater than 5")
    print("Program ended")
```

```python
# if-else Statement:
"""The if-else statement combines an additional block of code (the else block) 
       that runs when the if condition is false."""
x = 3
if x == 4:
    print("Yes")
else:
    print("No")
```

```python
#Nested if..else Statement:
#A nested if..else statement contains an inner if-else statement inside an outer if statement.
letter = "D"
if letter == "B":
    print("Letter is B")
else:
    if letter == "C":
        print("Letter is C")
    else:
        print("Letter isn't A, B, or C")

```

```python
# elif statements:
# It is used when we want to check for multiple conditions.
#Grading System:
score = 85
if score >= 90:
       grade = 'A'
elif score >= 80:
       grade = 'B'
elif score >= 70:
       grade = 'C'
elif score >= 60:
       grade = 'D'
else:
     grade = 'F'
print(f"Your grade is: {grade}")
   # Output: Your grade is: B
```

```python
# Ques7: Describe the different types of loops in Python and their use cases with examples


''' Ans:  In Python, loops are used to iterate over a sequence (like a list, tuple, dictionary, set, or string)
          or other iterable objects. There are two main types of loops in Python Each has its specific use
          cases and functionality.

     1. `for` Loop
         The `for` loop is used to iterate over a sequence (such as a list, tuple, dictionary, set, or string)
         or any other iterable object. It allows you to execute a block of code multiple times.

           Syntax:
           
             for item in sequence:
                 # code block

           Use Cases:
           
          a.  Iterating over a list of items.
            
          b.  Performing actions on elements in a sequence.
            
          c.  Traversing the keys or values in a dictionary. ''''
```

```python
#Example of for loop


 #1.Iterating Over a List:
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit,end=" ")
   # Output:
   # apple
   # banana
   # cherry
print() #for new line

 #2.Iterating Over a String:
for char in "hello":
    print(char,end=" ")
   # Output: h e l l o
print()
   

 #3.Iterating Over a Dictionary:
student_scores = {"Alice": 90, "Bob": 85, "Charlie": 92}
for student, score in student_scores.items():
    print(f"{student}: {score}")
   # Output:
   # Alice: 90
   # Bob: 85
   # Charlie: 92
   

 #4.Using `range()` Function:
for i in range(5):
    print(i)
   # Output:
   # 0
   # 1
   # 2
   # 3
   # 4

   
```

```python
''' 2. `while` Loop
     The `while` loop is used to execute a block of code as long as a condition is `True`. It is useful when
     the number of iterations is not known beforehand and depends on a condition.

       Syntax:
          while condition:
          # code block

       Use Cases:
         a. Repeating a block of code until a condition is met.
         
         b. Creating loops that depend on external factors or user input.
         
         c. Implementing complex loop conditions.''''
```

```python
#Examples of While loop:

#1.Basic `while` Loop:
count = 0
while count < 5:
    print(count)
    count += 1
   # Output:
   # 0
   # 1
   # 2
   # 3
   # 4
   
   

#2.Loop With User Input:
user_input = ""
while user_input.lower() != "exit":
    user_input = input("Enter something (type 'exit' to quit): ")
print(f"You entered: {user_input}")
   # The loop will keep running until the user types "exit".

   

 #3.Nested `while` Loop:
i = 0
while i < 3:
    j = 0
while j < 2:
    print(f"i = {i}, j = {j}")
    j += 1
    i += 1
   # Output:
   # i = 0, j = 0
   # i = 0, j = 1
   # i = 1, j = 0
   # i = 1, j = 1
   # i = 2, j = 0
   # i = 2, j = 1


```
